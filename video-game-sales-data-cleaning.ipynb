```python
import pandas as pd
df=pd.read_csv('/kaggle/input/video-game-sales/vgsales.csv')
```

```python
df.head()
```

```python
df=df.drop("Critic_Score",axis=1)
```

```python
df=df.drop("User_Score",axis=1)
```

```python
df=df.drop("NA_Sales",axis=1)
```

```python
df=df.drop("PAL_Sales",axis=1)
```

```python
df=df.drop("JP_Sales",axis=1)
```

```python
df=df.drop("Other_Sales",axis=1)
```

```python
df.head()
```

```python
print(df['Global_Sales'].isna().sum())
```

```python
df = df.dropna(subset=['Global_Sales'])
```

```python
print( len(df. index))
```

```python
df.head()
```

```python
df['Year'] = df['Year'].apply(lambda x: (int(str(x)[0:4])))
```

```python
df.head()
```

```python
df['Platform'].value_counts()
```

```python
platform_categories = {
    'Nintendo DS Family': ['DS', '3DS'],
    'PlayStation Family': ['PS2', 'PS3', 'PS4', 'PSV', 'PSP'],
    'Xbox Family': ['XB', 'X360', 'XOne'],
    'Nintendo Family': ['N64', 'GC', 'NES', 'SNES', 'Wii', 'WiiU', 'NS'],
    'Sega Family': ['GEN', 'SCD'],
    'Atari Family': ['2600'],
    'PC': ['PC', 'OSX'],
    'Other Handhelds': ['GBA', 'WS', 'NG', 'GG'],
    'Other Consoles': ['DC', 'PSN', 'XBL', 'GBC', 'VC', '3DO', 'PCE', 'PCFX', 'WW', 'Mob']
}

# Example usage
df['Platform_Category'] = df['Platform'].apply(lambda x: next((k for k, v in platform_categories.items() if x in v), 'Other_plat'))
```

```python
df.head()
```

```python
df['Developer'].value_counts()
```

```python
def developer(x):
    if 'Nintendo' in x:
        return 'Nintendo'
    elif 'Capcom' in x:
        return 'Capcom'
    elif 'Ubisoft' in x:
        return 'Ubisoft'
    elif 'Konami' in x:
        return 'Konami'
    elif 'EA' in x or 'Electronic Arts' in x:
        return 'EA'
    elif 'Namco' in x:
        return 'Namco'
    else:
        return 'Other_dev'
        
df['Developer_Category'] = df['Developer'].apply(developer)
```

```python
df.head()
```

```python
df['Publisher'].value_counts()
```

```python
def publisher(x):
    if 'Activision' in x:
        return 'Activision'
    elif 'Ubisoft' in x:
        return 'Ubisoft'
    elif 'Namco' in x or 'Bandai' in x:
        return 'Namco'
    elif x == 'Nintendo':
        return 'Nintendo'
    elif 'EA' in x or 'Electronic Arts' in x:
        return 'EA'
    elif 'Konami' in x:
        return 'Konami'
    else:
        return 'other_publisher'
df['Publisher_Category'] = df['Publisher'].apply(publisher)
```

```python
df=df.drop("Platform",axis=1)
df=df.drop("Publisher",axis=1)
df=df.drop("Developer",axis=1)
```

```python
df.head()
```

```python
df = df.drop(df.columns[0], axis=1)
df = df.drop(df.columns[0], axis=1)
```
